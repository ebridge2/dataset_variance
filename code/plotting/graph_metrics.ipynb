```python
from multiplot import feature_plot

import pickle
import os

import matplotlib.pyplot as plt
%matplotlib inline

font = {'weight' : 'bold',
        'size'   : 14}

import matplotlib
matplotlib.rc('font', **font)
```

## NNZ

```python
f_nnz = open('../../data/graphs/desikan_nnz.pkl')

nnz = pickle.load(f_nnz)['nnz']
f_nnz.close()
dnames = nnz.keys()

plt_nnz = feature_plot(nnz, dnames, 'Number of Non-zero Edges', plot_mode='scatter',
                       fig_outfile='../../figs/multi_nnz_desikan_scatter_new.png')
```

## Degree Distribution

```python
f_deg = open('../../data/graphs/desikan_degree.pkl')

degree = pickle.load(f_deg)['degree']
f_deg.close()
dnames = degree.keys()

plt_nnz = feature_plot(degree, dnames, 'Degree Distribution', plot_mode='hist', xlims = [0, 70],
                       ylab='Density', xlab= 'Vertex Degree', fig_outfile='../../figs/multi_degree_desikan_hist_new.png')
```

```python
f_ew = open('../../data/graphs/desikan_edgeweight.pkl')

ew = pickle.load(f_ew)['edgeweight']
f_ew.close()
dnames = ew.keys()

plt_nnz = feature_plot(ew, dnames, 'Edge Weight Distribution', plot_mode='hist', axis_scale='log', yscale=10000,
                       ylab='Density (E-4)', xlab='Edge Weight', fig_outfile='../../figs/multi_ew_desikan_hist_new.png')
```
