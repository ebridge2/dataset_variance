```python
from multiplot import feature_plot

import pickle
import os

import matplotlib.pyplot as plt
%matplotlib inline

font = {'weight' : 'bold',
        'size'   : 14}

import matplotlib
matplotlib.rc('font', **font)
```

## NNZ

```python
f_nnz = open('../../data/graphs/Talairach_nnz.pkl')

nnz = pickle.load(f_nnz)['nnz']
f_nnz.close()
dnames = nnz.keys()

plt_nnz = feature_plot(nnz, dnames, 'Number of Non-zero Edges (Talairach Atlas)', plot_mode='scatter',
                       fig_outfile='../../figs/graphs/talairach/multi_nnz_talairach_scatter.png')
```

## Degree Distribution

```python
f_deg = open('../../data/graphs/Talairach_degree.pkl')

degree = pickle.load(f_deg)['degree']
f_deg.close()
dnames = degree.keys()

plt_nnz = feature_plot(degree, dnames, 'Degree Distribution (Talairach Atlas)', plot_mode='hist', xlims = [0, 1105],
                       ylab='Density', xlab= 'Vertex Degree', fig_outfile='../../figs/graphs/talairach/multi_degree_talairach_hist.png')
```

```python
f_ew = open('../../data/graphs/Talairach_edgeweight.pkl')

ew = pickle.load(f_ew)['edgeweight']
f_ew.close()
dnames = ew.keys()

plt_nnz = feature_plot(ew, dnames, 'Edge Weight Distribution (Talairach Atlas)', plot_mode='hist', axis_scale='log', yscale=1000,
                       ylab='Density (E-3)', xlab='Edge Weight', fig_outfile='../../figs/graphs/talairach/multi_ew_talairach_hist.png')
```
